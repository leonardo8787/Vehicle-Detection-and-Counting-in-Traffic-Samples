```python
import glob
import os 

path_train = "/home/leonardocampos/.cache/kagglehub/datasets/ashfakyeafi/road-vehicle-images-dataset/versions/2/trafic_data/train"
path_valid = "/home/leonardocampos/.cache/kagglehub/datasets/ashfakyeafi/road-vehicle-images-dataset/versions/2/trafic_data/valid"

train_images_pattern = os.path.join(path_train, "images", "*.jpg")
train_labels_pattern = os.path.join(path_train, "labels", "*.txt")

valid_images_pattern = os.path.join(path_valid, "images", "*.jpg")
valid_labels_pattern = os.path.join(path_valid, "labels", "*.txt")

print("Listando arquivos...")
train_image_files = sorted(glob.glob(train_images_pattern))
train_label_files = sorted(glob.glob(train_labels_pattern))

valid_image_files = sorted(glob.glob(valid_images_pattern))
valid_label_files = sorted(glob.glob(valid_labels_pattern))

print(f"--- Treino ---")
print(f"Encontradas {len(train_image_files)} imagens (padrão: {train_images_pattern})")
print(f"Encontrados {len(train_label_files)} labels (padrão: {train_labels_pattern})")

print(f"\n--- Validação ---")
print(f"Encontradas {len(valid_image_files)} imagens (padrão: {valid_images_pattern})")
print(f"Encontrados {len(valid_label_files)} labels (padrão: {valid_labels_pattern})")
```

```python
import cv2

print("\n--- Exemplo de Leitura (Treino) ---")

if train_image_files and train_label_files:
    primeira_imagem_path = train_image_files[1]
    primeiro_label_path = train_label_files[1]

    try:
        imagem = cv2.imread(str(primeira_imagem_path))
        if imagem is not None:
            print(f"Imagem carregada: {primeira_imagem_path}")
            print(f"  -> Dimensões (Altura, Largura, Canais): {imagem.shape}")
        else:
            print(f"Erro ao carregar imagem: {primeira_imagem_path}")
    except Exception as e:
        print(f"Erro ao ler imagem {primeira_imagem_path}: {e}")

    try:
        with open(primeiro_label_path, 'r') as f:
            conteudo_label = f.read()
            print(f"Label carregado: {primeiro_label_path}")
            print(f"  -> Conteúdo do Label:\n{conteudo_label}")
    except Exception as e:
        print(f"Erro ao ler label {primeiro_label_path}: {e}")

else:
    print("Nenhum arquivo de imagem ou label encontrado nas pastas de treino.")
```

```python
from ultralytics import YOLO

model = YOLO('yolo12n.pt')

path_do_yaml = "/home/leonardocampos/.cache/kagglehub/datasets/ashfakyeafi/road-vehicle-images-dataset/versions/2/trafic_data/data_1.yaml"

print("Iniciando o treinamento...")
results = model.train(
    data=path_do_yaml,
    epochs=100,       
    imgsz=640,        
    batch=16          # Quantas imagens processar de uma vez (diminua se der erro de memória)
)

print("Treinamento concluído!")
print(f"Modelo e resultados salvos em: {results.save_dir}")
```
